# Chess Game

Board display:
- Function **display_board(board)**: displays the board.
- Function **init_board()**: returns a board, initialized the board and all the figures.
  - **board**: has the values of the board. 

Match
- Function **start_match()**:
  - init_board()
  - display_board()
  - While loop satisfies (!**is_end()** or !**is_draw()**), finishes the game.
- Function **is_end()**: returns boolean, checks if it the player has won.
  - king_can_move(): returns false
  - king_is_attacked(): returns true
  - **winner**: has the name of the winner.
- Function **is_draw()**: returns boolean, checks if it is a draw
  - King cannot move.
  - King made 50 "forced" movements.
  - Players repeated a move 3 times in a row.
  - Players agreed on a draw

Movements:
- Function **try_move(board, figure, end_square)**: tries  a move.
- Function **king_can_move(board, king_color)**: returns true if king has places to move.
- Function **king_is_attacked()**: returns true if king is being attacked.

Miscelánea:
- Function **get_piece**: gives an int and returns the looked for piece.
