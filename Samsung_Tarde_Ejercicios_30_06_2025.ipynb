### 1. Dadas dos listas: `keys` y `values`, crear un diccionario con ellas. En el caso de que aparezca más de una vez la misma clave, **se sumarán los valores**.

Con la siguientes listas:

```
keys = ["a", "b", "c", "b"]  
values = [1, 2, 3, 1] 
```

Se crearía el diccionario:

```
{"a": 1, "b": 3, "c": 3}
```

````python
import random

ALL_KEYS = "abcdefghijklmnopqrstuvwxyz"

keys = random.choices(ALL_KEYS, k=10)
values = [random.randint(1, 1000) for _ in range(10)]

def crea_dict(keys, values):
    res = {}
    for i in range(len(keys)):
        res[keys[i]] = res.get(keys[i], 0) + values[i]  
    return res

print("keys:", keys)
print("values:", values)
print(crea_dict(keys, values))
````

### 2. Dados dos diccionarios, realizar la `suma` de estos. La suma devolverá un diccionario con **todas** las claves que aparecen en los dos diccionarios. El valor para una clave será calculado como la suma de los valores, para la misma clave, en ambos diccionarios.


Ejemplo:

```
dict1 = {"a": 1, "b": 2}  
dict2 = {"b": 2, "c": 3, "d": 4}
```

Da como resultado:
```
{"a": 1, "b": 4, "c": 3, "d": 4}
```

````python
import random

ALL_KEYS = "abcdefghijklmnopqrstuvwxyz"

dict1 = {k: random.randint(1, 10) for k in random.choices(ALL_KEYS, k=10)}
dict2 = {k: random.randint(1, 10) for k in random.choices(ALL_KEYS, k=10)}

def suma(dict1, dict2):
    result = dict1.copy()
    for clave, valor in dict2.items():
        result[clave] = result.get(clave, 0) + dict2[clave]

    # Completar
    return result

print(dict1, "+", dict2, "=", suma(dict1, dict2))
````

### 12. Del fichero `el_quijote.txt`, obtener un diccionario en el cual la clave sean las palabras en **minúscula** y el valor sea la lista de líneas (posiciones) en las que aparece dicha palabra. El diccionario resultante se guardará en `el_quijote_poslist.json`. Reemplazar todos los caracteres no alfanuméricos por `' '`.

Ejemplo:

```
DON QUIJOTE DE LA MANCHA
Miguel de Cervantes Saavedra

PRIMERA PARTE
CAPÍTULO 1: Que trata de la condición y ejercicio del famoso hidalgo D. Quijote de la Mancha
```

daría:

```
{'don': [0], 'quijote': [0, 4], 'de': [0, 1, 4], 'la': [0, 4], 'mancha': [0, 4], 'miguel': [1], 'cervantes': [1], 'saavedra': [1], 'primera': [3], 'parte': [3], 'capítulo': [4], '1': [4], 'que': [4], 'trata': [4], 'condición': [4], 'y': [4], 'ejercicio': [4], 'del': [4], 'famoso': [4], 'hidalgo': [4], 'd': [4]}
```


````python
NON_ALNUM = [
    chr(i) for i in range(256)
    if not chr(i).isalnum()
]

d_palabras = {}
count = 1
with open('el_quijote.txt', 'r', encoding="utf-8") as f:
    for l in f:
        s = l
        for i in NON_ALNUM:
            s = s.replace(i, " ")
        s = s.split()
        for word in s:
            if word.islower():
                if word not in d_palabras:
                    d_palabras[word] = [count]
                else:
                    if d_palabras[word][-1] != count:
                        d_palabras[word].append(count)
        count += 1

with open('el_quijote_poslist.json', 'w', encoding="utf-8") as f:
    f.write(json.dumps(d_palabras))
    
print(d_palabras)
````
